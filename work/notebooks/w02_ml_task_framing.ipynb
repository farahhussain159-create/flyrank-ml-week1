## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
baseline_precision_at_50 = 0.24
model_precision_at_50 = 0.74
lift = model_precision_at_50 / baseline_precision_at_50
print(f"Baseline: {baseline_precision_at_50}, Model: {model_precision_at_50}, Lift: {lift:.1f}x")
```

A hand-written rule (the repo's baseline score) only reaches Precision@50 = 0.24.
A trained model on the same task reaches Precision@50 ≈ 0.74 — roughly a 3x lift.
Declining performance comes from many signals interacting non-linearly (traffic drop,
ranking shift, CTR change, seasonality), which a simple if-else rule can't capture,
but a model can learn from the data.
